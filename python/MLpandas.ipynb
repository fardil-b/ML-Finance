Creating a new Price Increase series (Price Target - Current Price)

```python
csv_dataframe['Price Increase'] = csv_dataframe['Price Target'] - csv_dataframe['Current Price']
```

```python
csv_dataframe.head(5)
```

```python
csv_dataframe['Return'] = csv_dataframe['Price Increase'] / csv_dataframe['Current Price']
```

```python
csv_dataframe.Return.()
```

```python
csv_dataframe.Return.describe()
```

```python
boolean_mask = csv_dataframe.Return > 0.15
```

```python
boolean_mask
```

This __csv_dataframe[boolean_mask]__ will show only stocks with return greater than 0.15

```python
csv_dataframe[boolean_mask]
```

This ~ operator shows the inverse, that is stock that are < 0.15

```python
csv_dataframe[~boolean_mask]
```

Dsiplaying only technology company
